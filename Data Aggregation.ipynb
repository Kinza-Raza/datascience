```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb
```

```python
file = pd.read_csv("World_Happiness_2015.csv")
file
```

```python
region = file["Region"].unique()
region #array not dataframe
```

```python
#Finding mean of happiness score region wise
region_mean = {}
for i in region:
    df = file[file["Region"] == i]
    hp_score = df["Happiness Score"]
    sc_mean = hp_score.mean()
    region_mean[i] = sc_mean
region_mean
```

```python
#Alternative shorter way of doing same procedure as above
reg = file.groupby("Region") #dataframe created region wise but it wont print
score = reg["Happiness Score"]
score.mean()
```

```python
rank = reg["Happiness Rank"]
rank.mean() #finding mean of happiness rank region wise
```

```python
r_mean = score.agg(["mean","max", np.min]) #finding mean, max and min value region wise, passing multiple functions in list
r_mean
```

```python
hpscore_mean = file.pivot_table(index = "Region", values="Happiness Score") #by default it shows mean as aggfunc=mean
#hpscore_mean
hpscore_agg = file.pivot_table(index = "Region", values="Happiness Score", aggfunc=[np.min,np.max]) #specifying min,max
hpscore_agg
```

```python
#plotting graph of mean happiness score region wise 
fig, ax = pt.subplots(figsize=(12,8))
sb.barplot(x=hpscore_mean.index, y=hpscore_mean["Happiness Score"], data=hpscore_mean, orient="v")
pt.xticks(rotation=90)
```
