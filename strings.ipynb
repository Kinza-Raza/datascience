```python
pattern = r"[Nn]ational accounts"
```

```python
merged[merged["SpecialNotes"].str.contains(pattern, na=False)]["SpecialNotes"].head() #displays all records where national accounts appears
```

```python
ptn = r"([1-2][0-9][0-9][0-9])"
#ptn = r"([1-2][0-9]{3})" 
merged[merged["SpecialNotes"].str.contains(ptn, na=False)].head()
```

```python
merged["SpecialNotes"].str.extract(ptn).head(10)
```

str.extract returns records direcly, pattern must be enclosed in () but contains returns bool value

if anything repeated in pattern, write number of times its repeated in {}

```python
ptn2 = r"([1-2][0-9][0-9][0-9])-?([1-2][0-9][0-9][0-9])?"
merged["SpecialNotes"].str.extract(ptn2).tail()
```

```python
pa = r"([1-2][0-9][0-9][0-9])(/)?([0-9][0-9])?" # / is in separate capture group
merged["IE Survey"].str.extract(pa) 
```
