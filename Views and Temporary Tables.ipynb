```python
import pandas as pd
import numpy as np 
import sqlite3 as sql
import matplotlib.pyplot as plt 
```

```python
db = "chinook.db"
con = sql.connect(db)
def run(q):
    return pd.read_sql_query(q,con)
```

```python
q = """select name,type from sqlite_master where type='table'"""
run(q)
```

```python
q = """select pl.playlist_id, pl.name, count(pt.track_id) "no. of tracks", sum((tr.milliseconds)/1000) "length in seconds"
        from playlist pl inner join playlist_track pt 
        on pl.playlist_id = pt.playlist_id
        inner join track tr
        on pt.track_id = tr.track_id
        group by 1"""
run(q)
```

```python
q = """create View customers_90 as 
        select c.*
        from customer c inner join invoice i
        on c.customer_id = i.customer_id
        group by 1
        having sum(i.total) > 90
        """      
run(q)
```

```python
q = """select * from customer_90"""
run(q)
```

```python
q = """select name,type from sqlite_master where type='view'"""
run(q)
```

```python
q = """create View usa as
       select * from customer 
       where country='USA'
        """
run(q)
```

```python
q = """select * from usa limit 5"""
run(q)
```

```python
q = """select * from customers_90 union 
        select * from usa limit 5"""
run(q)
```

By creating temporary table, joining of temp. table with existing table 'employee'.
Finding employees who deal with customers located in USA AND customers having total purchases greater than 90

```python

q = """With customer_usa_90 as (select * from customers_90
        intersect select * from usa)
        
        select e.first_name||' '||e.last_name Name, count(c.customer_id) "no. of customers"
        from employee e inner join customer_usa_90 c 
        on e.employee_id = c.support_rep_id
        group by 1
        having e.title = "Sales Support Agent"
        """
run(q)
```

Intersect and Union only work when number of columns are same in both queries and data type should be same

Except is used to show records of first query not present in the second query!

```python
q = """ with cust_total as (select c.customer_id, c.first_name||' '||c.last_name Name,sum(i.total) "Total"
        from customer c inner join invoice i
        on c.customer_id = i.customer_id group by 1 )

        select c.country 'Country', ct.Name, max(ct.Total)
        from customer c left join cust_total ct on 
        c.customer_id = ct.customer_id
        group by 1
        order by 1
        """
run(q)
```
