```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
```

```python
file = pd.read_csv("unrate.csv")
```

```python
fig = mp.figure(figsize = (15,10))
plot1 = fig.add_subplot(2,1,1) #first number indicates rows, second indicates column,third indicates number of plots
plot2 = fig.add_subplot(2,1,2)
plot1.plot(file["DATE"].head(12), file["VALUE"].head(12))
plot2.plot(file["DATE"][12:24], file["VALUE"][12:24])
mp.xticks(rotation = 90) #doesnt change label rotation for first plot
mp.show()
```

```python
fig = mp.figure(figsize = (15,10))
plot1 = fig.add_subplot(2,1,1) #first number indicates rows, second indicates column,third indicates number of plots
plot2 = fig.add_subplot(2,1,2)
plot1.plot(file["DATE"].head(12), file["VALUE"].head(12))
plot2.plot(file["DATE"][12:24], file["VALUE"][12:24])
mp.xticks(rotation = 90) #doesnt change label rotation for first plot
mp.show()
```

```python
#PRINTING SUBPLOTS FOR YEARS 1948 - 1952
fig = mp.figure(figsize = (15,10))
for i in range(0,5):
    ax = fig.add_subplot(5,1,i+1)
    ax.plot(file["DATE"][i*12:(i+1)*12], file["VALUE"][i*12:(i+1)*12])
mp.show()
```

```python
import datetime as dt
file["DATE"] = pd.to_datetime(file["DATE"])
file["months"] = file["DATE"].dt.month
```

```python
mp.plot(file["months"][0:12], file["VALUE"][0:12], label = "1948", c = "black") #use c to change colour of line
mp.plot(file["months"][12:24], file["VALUE"][12:24] , label = "1949")
mp.legend(loc = "upper left") #key which gives labels
```

```python
x = 1948
colors = ["red", "green", "blue", "orange", "brown"]
for i in range(0,5):
    start_index = 12*i
    end_index = (i+1)*12
    lbl = x+i
    mp.plot(file["months"][start_index:end_index], file["VALUE"][start_index:end_index], label = lbl, c = colors[i])
mp.legend(loc = "upper left")
mp.xlabel("Months")
mp.ylabel("Unemployment Rate")
mp.title("Graph of Unemployment from 1948-1952")
mp.show()
```
