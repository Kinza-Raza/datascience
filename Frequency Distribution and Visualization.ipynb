```python
intervals = pd.interval_range(0,600,freq=100)
intervals
```

```python
pts_freq = pd.Series(np.zeros(6),index = intervals)
pts_freq
```

```python
for i in file["PTS"]:
    #print(i)
    for j in intervals:
        
        if i in j:
            pts_freq.loc[j] += 1
            break
```

```python
file["Experience"].value_counts(dropna = False)
```

```python
exp = pd.to_numeric(file["Experience"], errors = "coerce")
exp
```

```python
def rank(x):
    if pd.isnull(x):
        return "Rookie"
    elif 1<= x <= 5:
        return "Less Experienced"
    elif 5< x <= 10:
        return "Experienced"
    else:
        return "Very Experienced"
```

```python
file["exp_ordinal"] = exp.apply(rank)
file["exp_ordinal"].head(7)
```

```python
#exp = file["Experience"].replace("R","0")
```

```python
file["exp_ordinal"].value_counts()
```

```python
file["exp_ordinal"].value_counts().iloc[[2,0,1,3]].plot.bar(rot = 45)
plt.title("WNBA Players Experience", y = 1.1) #y creates some distance b/w title and plot
plt.show()
```

```python
file["exp_ordinal"].value_counts().iloc[[2,0,1,3]].plot.barh()
plt.title("WNBA Players Experience", y=1.1)
plt.show()
```

Horizontal bar graph helps in better understanding

```python
fig, ax = plt.subplots(figsize = (5,5))
file["exp_ordinal"].value_counts().iloc[[2,0,1,3]].plot.pie(autopct = "%.1f%%", startangle=-90,
                                                           counterclock = False) #autopct gives percentage
plt.title("WNBA Players Experience")
plt.ylabel("")
plt.show()
```
