```python
parameter_pt = file["PTS"].mean()
parameter_pt
```

```python
sample_mean = []
for i in range(1,101):
    sample = file["PTS"].sample(10,random_state=i)
    s_mean = sample.mean()
    sample_mean.append(s_mean)
sample_mean
```

*Data has 4 kinds of variables*

4 ways to measure variables - Scale of Measurement / Level of Measurement:
- Nominal -- Qualitative 
- Ordinal -- Quantitative (bar graph, pie chart)
- Interval -- Quantitative
- Ratio -- Quantitative (kde plot, histogram)

- in nominal you can only tell the data is not repeating. the direction nor the size is unknown. eg. name,gender,id number
- in ordinal the data is known to be unique, and direction. for eg. tall,short
- in interval, all three factors are known. 'zero' has no importance. eg. temp in Celsius
- in ratio, all three factors are known. 'zero' indicates absence of values. eg. height,width
