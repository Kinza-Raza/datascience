```python
import pandas as pd

file = pd.read_csv("laptops.csv",encoding = "Windows-1251")
file
```

```python
def clean_col(item):
    item = item.strip() #stripl for stripleft and stripr for stripright by default for spaces
    item = item.replace("(","")
    item = item.strip(")")
    item = item.lower()
    item = item.replace("operating system","os")
    if " " in item:
        item = item.replace(" ","_")
    return item
    
    
clean_col('Operating System Version')
clean_col('Price (Euros)')
```

```python
clean_column = []
for i in file.columns:
    clean = clean_col(i)
    clean_column.append(clean)
clean_column
```

```python
file.columns = clean_column
file.columns
```
