```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as mt 
```

```python
#1.READ DATA INTO PANDAS
file = pd.read_csv("recent-grads.csv")
file
```

```python
#2.CHECK DATA INFO
file.info()
```

```python
#3.PRINT FIRST FIVE ROWS
file.head(5)
```

```python
#4.REMOVE ALL ROWS WHERE DATA IS MISSING
file = file.dropna()
file.info()
```

```python
#5.Generate scatter plot between SAMPLE SIZE and MEDIAN
ax = file.plot(x="Sample_size", y="Median", kind="scatter", title="Scatter Plot", figsize=(5,5))
ax
```

```python
#6.Generate histogram of SAMPLE SIZE
ht = file["Sample_size"].plot(kind="hist", bins = 25, title="Sample Size")
ht
```

```python
#7.Generate histogram of SAMPLE SIZE, MEDIAN, EMPLOYED and FULL TIME 
name = ["Sample_size", "Median", "Employed", "Full_time"]
fig = mt.figure(figsize = (8,10))
for i in range(4):
    qwe = fig.add_subplot(2,2,i+1)
    qwe.hist(x=file[name[i]], bins=20)
    qwe.set_xlabel(name[i])
```

```python
#8.Generate bar diagram of first ten rows
bar = file.head(10).plot(kind="barh",x="Major", y="ShareWomen")
bar.set_ylabel("ShareWomen")
```
