```python
#1.Read data in pandas
import pandas as pd

file = pd.read_csv("autos.csv",encoding="windows-1251")
```

```python
#2.Review data using info()
file.info()
```

```python
#2.Review data using head()
file.head(7)
```

```python
#3.Streamline columns name using same pattern
streamline = {"dateCrawled":"date_crawled","name":"name","seller":"seller","offerType":"offer_type","price":"price",
              "abtest":"abtest","vehicleType":"vehicle_type","yearOfRegistration":"year_of_registration","gearbox":"gearbox",
              "powerPS":"power_ps","model":"model","odometer":"odometer","monthOfRegistration":"month_of_registration",
              "fuelType":"fuel_type","brand":"brand","notRepairedDamage":"not_repaired_damage","dateCreated":"date_created",
              "nrOfPictures":"nr_of_pictures","postalCode":"postal_code","lastSeen":"last_seen"}
file.columns = file.columns.map(streamline)
file.columns
```

```python
#4.Use .describe() to understand stats of dataframe
file.describe()
```

```python
#5.Delete num_photos, seller, and offer_type columns
file.drop(["offer_type","nr_of_pictures","seller"],axis = 1)
```

Columns "seller", "offer_type", "nr_of_pictures" deleted as repetitive data in each record

```python
#6.Convert price to int
file.price = file.price.str.strip("$")
file.price = file.price.str.replace(",","").astype(int)
file.price.dtype
```

```python
#6.Convert odometer to int
file.odometer = file.odometer.str.replace(",","")
file.odometer = file.odometer.str.strip("km").astype(int)
file.odometer.dtype
```

```python
#7.Rename odometer to "odometer_km"
file["odometer_km"] = file["odometer"]
file.odometer_km
```

```python
#8.Keep data where price greater than equal to 1 and less than equal to 350000
file[file["price"].between(1,350000)]
```

```python
#9.Keep data where year of registration not less than 1900 and not greater than 2016
file[file["year_of_registration"].between(1900,2016)]
```

```python
#10.Finding top 5 brands based on frequency
file["brand"].value_counts()
```

```python
#10.Calculating mean price and mean mileage of top 5 brands
top_5 = ["volkswagen","opel","bmw","mercedes_benz","audi"]
mean_price = {}
mean_mileage = {}
for i in top_5:
    row = file[file["brand"] == i]
    price = row["price"].mean()
    mean_price[i] = price
    mileage = row["odometer_km"].mean()
    mean_mileage[i] = mileage
print(mean_price)
print(mean_mileage)
```

```python
#11.Converting both dictionaries to Series
mean_price_series= pd.Series(mean_price)
mean_mileage_series= pd.Series(mean_mileage)
print(mean_price_series)
print(mean_mileage_series)
```

```python
#12.Converting mean price series to dataframe
meanprice_dataframe = pd.Series.to_frame(mean_price_series)
meanprice_dataframe
```

```python
#Converting mean mileage series to dataframe
meanmileage_dataframe = pd.Series.to_frame(mean_mileage_series)
meanmileage_dataframe
```
