```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as bmap
```

```python
file = pd.read_csv("titanic.csv")
file
```

```python
cols = ["Survived", "Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
f=file[cols]
```

```python
f=f.dropna()
f.info()
```

```python
#data = file[cols]
sb.distplot(f["Age"])
```

```python
sb.kdeplot(f["Age"], shade=True) #kdeplot removes bars
sb.despine() #despine removes top and right by default 
sb.set_style("ticks")
```

```python
sb.kdeplot(f["Age"], shade=True) 
sb.despine()
plt.xlabel("Age")
```

```python
g = sb.FacetGrid(f, col="Survived", height=4) #in col write column name of first condition
g.map(sb.kdeplot,"Age", shade=True)
```

g = sb.FacetGrid(f, col="Survived", height=4)
g.map(sb.kdeplot,"Age", shade=True)

```python
g = sb.FacetGrid(f, col="Pclass", height=4)
g.map(sb.kdeplot,"Age", shade=True)
```

```python
g = sb.FacetGrid(f, col="Pclass", row = "Survived",  height=4)
g.map(sb.kdeplot,"Age", shade=True)
```

```python
g = sb.FacetGrid(f, col="Pclass", row = "Survived", hue="Sex",  height=4)
g.map(sb.kdeplot,"Age", shade=True)
g.add_legend()
g.despine(bottom = False, left = True)
```
