```python
q = """select c.first_name, inv.total from customer c inner join invoice inv
        on c.customer_id = inv.customer_id"""
run(q)
```

```python
#Contents on Invoice 4
q = """select tr.track_id, tr.name Track_name, ar.name Artist, md.name Track_type, inv.unit_price, inv.quantity
        from track tr inner join media_type md on tr.media_type_id = md.media_type_id inner join invoice_line inv 
         on tr.track_id = inv.track_id inner join album alb on tr.album_id = alb.album_id
         inner join artist ar on alb.artist_id = ar.artist_id where inv.invoice_id= '4'
    """
run(q)
```

```python
q = "select * from invoice_line"
run(q)
```

```python
q = """select art.name Artist, alb.title Album, count(inv.quantity) Sold from
        artist art inner join album alb on art.artist_id = alb.artist_id
        inner join track tr on alb.album_id = tr.album_id
        inner join invoice_line inv on tr.track_id = inv.track_id group by 2 order by 3 desc limit 5
"""
run(q)
```

```python
q = """select e.first_name||' '||e.last_name Emp_name, e.title Emp_title, m.first_name||' '||m.last_name Manager_name,
        m.title Manager_title
        from employee e left join
    employee m on e.reports_to = m.employee_id
    """
run(q)
```

```python
q = """select phone, first_name, last_name from customer where first_name like '%jen%'"""
run(q)
```

```python
q = """select c.first_name||' '||c.last_name Name, count(inv.invoice_date) number_of_purchases, sum(inv.total) Total
        from customer c inner join invoice inv on c.customer_id = inv.customer_id group by Name order by Total desc limit 5
"""
run(q)
```
