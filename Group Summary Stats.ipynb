```python
import pandas as pd
import numpy as np 
import sqlite3 as sql
```

```python
db = "jobs.db"
con = sql.connect(db)
```

```python
def run(q):
    return pd.read_sql_query(q,con)
```

```python
q = """select name,type from sqlite_master where type='table'"""
run(q)
```

```python
q = """select Major_Category, avg(ShareWomen) Average
        from recent_grads
        group by Major_Category
        Order by Average desc
        """
run(q)
```

```python
#alternative query 
q = """select Major_Category, avg(ShareWomen) Average
        from recent_grads
        group by 1
        Order by 2 desc
        """
run(q)
```

```python
q = """select Major_Category, avg(Low_wage_jobs)/avg(Total) share_wage
         from recent_grads group by Major_Category
       having share_wage > 0.1
        Order by share_wage desc
        """
run(q)
```
