```python
fifa = pd.read_csv("fifa.csv", index_col = "Date")
fifa
```

```python
fifa["ARG"] = fifa["ARG"].astype(int)
fifa["BRA"] = fifa["BRA"].astype(int)
fifa["ESP"] = fifa["ESP"].astype(int)
fifa["FRA"] = fifa["FRA"].astype(int)
fifa["GER"] = fifa["GER"].astype(int)
fifa["ITA"] = fifa["ITA"].astype(int)
fifa.info()
```

```python
fig, ax= plt.subplots(figsize=(10,10))
sb.lineplot(data=fifa["BRA"], label="Brazil") #plotting one column's data
sb.set_style("dark") #(1)darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"
```

```python
fl = pd.read_csv("World_Happiness_2016.csv")
fl
```

```python
fl["Health"] = fl["Health (Life Expectancy)"]
fl["Economy"] = fl["Economy (GDP per Capita)"]
cols = ["Country", "Happiness Score", "Economy", "Health"]
sb.boxplot(data=fl[cols]) #based on all countries
```

# A long-form DataFrame, in which case the x, y, and hue variables will determine how the data are plotted.
# A wide-form DataFrame, such that each numeric column will be plotted.(when x,y not mentioned)

```python
f = fl[cols]
c = f["Country"].isin(["Denmark", "Iceland"])
a = f[c] 
a
```

```python
sb.boxplot(data=f[c], x=a["Country"], y="Economy") #country vs economy
```

```python
fig, ax= plt.subplots(figsize=(8,16))
sb.boxplot(data=f[0:101], saturation=1) #boxplot of first 100 countries in rank based on happiness, economy, health
```
