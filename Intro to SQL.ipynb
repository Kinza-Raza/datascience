```python
import pandas as pd
import sqlite3 as sql
```

```python
con = sql.connect("jobs.db")
q = "select * from sqlite_master where type='table'"
pd.read_sql_query(q,con)
```

```python
def run_q(q):
    return pd.read_sql_query(q,con)
```

```python
q = "select * from recent_grads limit 5" #limit works as head 
run_q(q)
```

```python
q = "select * from recent_grads where ShareWomen > 0.5 limit 6;"
run_q(q)
```

```python
q = """select Rank,Major,ShareWomen,Unemployment_rate from recent_grads where Major_category = 'Engineering'
and (ShareWomen < 0.5 or Unemployment_rate < 0.05) order by Major asc limit 10"""
run_q(q)
```

Sum gives answer in INT while total gives in FLOAT
