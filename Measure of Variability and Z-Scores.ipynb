```python
dist = []
c = [1,1,1,1,1,1,21]
mean = np.mean(c)
for i in c:
    d = abs(i-mean)
    dist.append(d)
avg = sum(dist)/len(c)
avg   
```

```python

def standard_deviation(array):
    sq_dist = []
    for i in array:
        sqd = (i - mean)**2
        sq_dist.append(sqd)
    var = sum(sq_dist)/len(c)
    sd = var**(1/2)
standard_deviation(c)
```

```python
file = pd.read_csv("AmesHousing_1.txt", sep = "\t")

pd.options.display.max_columns = 82
file.head(3)
```

```python
sp_mean = file["SalePrice"].mean()
sp_mean
```

```python
#standard deviation of population
sp_std = file["SalePrice"].std()
sp_std
```

```python
sample = []
for i in range(5000):
    s = file["SalePrice"].sample(10, random_state=i)
    sd = s.std()
    sample.append(sd)
sample
```

```python
series = pd.Series(sample)
series
```

```python
series.hist()
plt.axvline(sp_std, color = 'black')
plt.xlim(20000,225000)
```

```python
s1 = file["SalePrice"].sample(1, random_state = 0)
s1
```

```python
file["SalePrice"].hist()
plt.axvline(sp_mean, color = 'black', label="mean")
plt.axvline(sp_mean+sp_std, color = 'red', label="limit")
plt.axvline(s1.values, color = 'green', label="sample")
plt.legend()
```

```python
z_score = ((220000 - sp_mean)/sp_std)
z_score
```

Z-Score tells how much standard deviation is there from a particular sample
